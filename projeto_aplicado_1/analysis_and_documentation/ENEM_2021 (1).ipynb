```python
# importando bibiliotecas e lendo o dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


db = pd.read_csv('enem_2021.csv', delimiter = ';')
```

```python
## calculando a média, mediana, moda e a nota máxima de cada prova do ENEM 2021

# Média
media_cn = round(db['NU_NOTA_CN'].mean(), 2)
media_ch = round(db['NU_NOTA_CH'].mean(), 2)
media_lc = round(db['NU_NOTA_LC'].mean(), 2)
media_mt = round(db['NU_NOTA_MT'].mean(), 2)
media_redacao = round(db['NU_NOTA_REDACAO'].mean(), 2)

# Mediana
mediana_cn = round(db['NU_NOTA_CN'].median(), 2)
mediana_ch = round(db['NU_NOTA_CH'].median(), 2)
mediana_lc = round(db['NU_NOTA_LC'].median(), 2)
mediana_mt = round(db['NU_NOTA_MT'].median(), 2)
mediana_redacao = round(db['NU_NOTA_REDACAO'].median(), 2)

# Moda
moda_cn = db['NU_NOTA_CN'].mode()[0]
moda_ch = db['NU_NOTA_CH'].mode()[0]
moda_lc = db['NU_NOTA_LC'].mode()[0]
moda_mt = db['NU_NOTA_MT'].mode()[0]
moda_redacao = db['NU_NOTA_REDACAO'].mode()[0]

# Máximo
max_cn = db['NU_NOTA_CN'].max()
max_ch = db['NU_NOTA_CH'].max()
max_lc = db['NU_NOTA_LC'].max()
max_mt = db['NU_NOTA_MT'].max()
max_redacao = db['NU_NOTA_REDACAO'].max()

# Imprimindo os resultados
print(f"CN: MÉDIA: {media_cn}, MEDIANA: {mediana_cn}, MODA: {moda_cn}, MAX: {max_cn}")
print(f"CH: MÉDIA: {media_ch}, MEDIANA: {mediana_ch}, MODA: {moda_ch}, MAX: {max_ch}")
print(f"LC: MÉDIA: {media_lc}, MEDIANA: {mediana_lc}, MODA: {moda_lc}, MAX: {max_lc}")
print(f"MT: MÉDIA: {media_mt}, MEDIANA: {mediana_mt}, MODA: {moda_mt}, MAX: {max_mt}")
print(f"Redação: MÉDIA: {media_redacao}, MEDIANA: {mediana_redacao}, MODA: {moda_redacao}, MAX: {max_redacao}")

```

```python
## calculando a variancia e o desvio padrão de cada prova do ENEM 2021


variancia_cn = round(db['NU_NOTA_CN'].var(), 2)
variancia_ch = round(db['NU_NOTA_CH'].var(), 2)
variancia_lc = round(db['NU_NOTA_LC'].var(), 2)
variancia_mt = round(db['NU_NOTA_MT'].var(), 2)
variancia_redacao = round(db['NU_NOTA_REDACAO'].var(), 2)

desvio_padrao_cn = round(db['NU_NOTA_CN'].std(), 2)
desvio_padrao_ch = round(db['NU_NOTA_CH'].std(), 2)
desvio_padrao_lc = round(db['NU_NOTA_LC'].std(), 2)
desvio_padrao_mt = round(db['NU_NOTA_MT'].std(), 2)
desvio_padrao_redacao = round(db['NU_NOTA_REDACAO'].std(), 2)

print("Prova CN:")
print(f"Variância: {variancia_cn}")
print(f"Desvio Padrão: {desvio_padrao_cn}")

print("\nProva CH:")
print(f"Variância: {variancia_ch}")
print(f"Desvio Padrão: {desvio_padrao_ch}")

print("\nProva LC:")
print(f"Variância: {variancia_lc}")
print(f"Desvio Padrão: {desvio_padrao_lc}")

print("\nProva MT:")
print(f"Variância: {variancia_mt}")
print(f"Desvio Padrão: {desvio_padrao_mt}")

print("\nRedação:")
print(f"Variância: {variancia_redacao}")
print(f"Desvio Padrão: {desvio_padrao_redacao}")
```

```python
# criando histogramas com as notas de cada prova

colunas = ['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']

hist_config = {
    'bins': 10,         
    'edgecolor': 'black'
}

# Loop para criar histogramas separados

for coluna in colunas:
    plt.figure(figsize=(6, 4)) 
    plt.hist(db[coluna], **hist_config)
    plt.title(f'Histograma de {coluna}')
    plt.xlabel('Nota')
    plt.ylabel('Alunos')
    plt.show()




```

```python
## calculando a média das provas dos estudantes de escola rural e urbana

urbano = (db['TP_LOCALIZACAO_ESC'] == 1.0)
rural = (db['TP_LOCALIZACAO_ESC'] == 2.0)

media_urbano = db.loc[urbano, ['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean()

media_rural = db.loc[rural, ['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean()

print(f"Média das provas dos estudantes de escola localizada na área urbana: {round(media_urbano.mean(),2)}")
print(f"Média das provas dos estudantes de escola localizada na área rural: {round(media_rural.mean(),2)}")
```

```python
## calculando a média das provas dos estudantes da UF mais populosa (SP) e menos populosa (RR) do Brasil

sp = (db['SG_UF_ESC'] == 'SP')
rr = (db['SG_UF_ESC'] == 'RR')

media_sp = db.loc[sp, ['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean()

media_rr = db.loc[rr, ['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean()

print(f"Média das provas dos estudantes de SP: {round(media_sp.mean(),2)}")
print(f"Média das provas dos estudantes de RR: {round(media_rr.mean(),2)}")
```

```python
## calculando a média das provas dos estudantes de escola pública e privada

publica = (db['TP_ESCOLA'] == 2)
privada = (db['TP_ESCOLA'] == 3)

media_publica = db.loc[publica, ['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean()

media_privada = db.loc[privada, ['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean()

print(f"Média das provas dos estudantes de escola pública: {round(media_publica.mean(),2)}")
print(f"Média das provas dos estudantes de privada: {round(media_privada.mean(),2)}")
```

```python
# Mapeamento entre UF e regiões
mapeamento_uf_regiao = {
    'Norte': ['AC', 'AP', 'AM', 'PA', 'RO', 'RR', 'TO'],
    'Nordeste': ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE'],
    'Centro-Oeste': ['DF', 'GO', 'MT', 'MS'],
    'Sudeste': ['ES', 'MG', 'RJ', 'SP'],
    'Sul': ['PR', 'RS', 'SC']}
    
def mapear_uf_para_regiao(uf):
    for regiao, ufs in mapeamento_uf_regiao.items():
        if uf in ufs:
            return regiao

# Criando uma nova coluna 'Regiao' com base na coluna 'UF'
db['Regiao'] = db['SG_UF_ESC'].apply(mapear_uf_para_regiao)
```

```python
## calculando a média das provas dos estudantes por região do Brasil

media_notas_por_regiao = db.groupby('Regiao')[['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean()

display(round(media_notas_por_regiao.mean(axis = 1)))
```

```python
## Coluna de médias 

db['NT_MEDIA'] = db[['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean(axis=1)

## Mapa de correlações

plt.rcParams['figure.figsize'] = (18, 10)

corr = db.corr()
matrix = np.triu(corr)

sns.heatmap(corr, annot = False, cmap = "viridis", mask = matrix)
plt.title('Heatmap for the Dataset', fontsize = 20)
plt.show()
```
