```python
# importando bibiliotecas e lendo o dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


db = pd.read_csv('ed_basica_2020.csv', delimiter = ';')
```

```python
## excluindo a coluna de indice 'Unnamed: 0'

db = db.drop(columns = ['Unnamed: 0'])
```

```python
# gráfico de pizza para analisar se há exame para ingressar na escola, onde 0 = não, 1 = sim e 9 = não informado

exame_para_ingressar = db['IN_EXAME_SELECAO'].value_counts()

plt.figure(figsize=(3, 3))
plt.pie(exame_para_ingressar, labels=exame_para_ingressar.index, autopct='%1.1f%%', startangle=140)
plt.title('Há exame para ingressar na escola?')

plt.axis('equal')
plt.show()
```

```python
# gráfico de pizza para analisar se há internet na escola, onde 1 = sim e 0 = não

acesso_internet = db['IN_INTERNET_ALUNOS'].value_counts()

plt.figure(figsize=(3, 3))  # Ajuste o tamanho da figura, se necessário
plt.pie(acesso_internet, labels=acesso_internet.index, autopct='%1.1f%%', startangle=140)
# Adicione um título
plt.title('Candidato tem acesso a internet na sua residência?')

plt.axis('equal')  # Isso garante que o gráfico seja uma pizza perfeita
plt.show()
```

```python
# gráfico de barras para analisar a distribuição da concentração de escolas por UF

frequencia = db['SG_UF'].value_counts().reset_index()
frequencia.columns = ['Valor', 'Frequência']
plt.figure(figsize=(10, 6))
plt.bar(frequencia['Valor'], frequencia['Frequência'], color='blue')
plt.title('Quantidade de escolas por UF')
plt.xlabel('Quantidade de escolas')
plt.ylabel('UF')
```
